# Precisamos tratar a base um pouco, tirando NAs primeiramente

```python
data = data.dropna()
```

# Agora precisamos colocar dummies para as variÃ¡veis categÃ³ricas ( Type, Method, e Regionname)

```python
prefixos = {'Type':'Tipo', 'Method':'MÃ©todo', 'Regionname':'RegiÃ£o'}

data = pd.get_dummies(data, columns = ['Type', 'Method','Regionname'], drop_first = True, prefix = prefixos)
```
