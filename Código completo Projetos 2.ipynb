# Ridge

```python
from sklearn.linear_model import Ridge
seed = 42

y = data['Price']
X = data.loc[:, data.columns != 'Price']
```

```python
from sklearn import preprocessing

minmax = preprocessing.MinMaxScaler()
X = minmax.fit_transform(X)
y = minmax.fit_transform(y.values.reshape(-1,1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = seed)
```

```python
from sklearn.model_selection import GridSearchCV


#Definindo o modelo, no caso, Ridge otimizado com cross validation
regridge = Ridge()
cvridge = GridSearchCV(regridge, param_grid = {'alpha' : np.linspace(0, 3, 11)})


#Fitando o modelo
cvridge.fit(X_train, y_train)
cvridge.best_params_
```

```python
regridgeotimo = Ridge(alpha = 0.6)
regridgeotimo.fit(X_train, y_train)
regridgeotimo.score(X_test, y_test)
```

# Lasso

```python
from sklearn.linear_model import Lasso
```

```python
reglasso = Lasso()
cvlasso = GridSearchCV(reglasso, param_grid = {'alpha' : np.linspace(1, 3, 11)})

cvlasso.fit(X_train, y_train)
cvlasso.best_params_
```

```python
reglassootimo = Lasso(alpha = 1)
reglassootimo.fit(X_train, y_train)
reglassootimo.score(X_test, y_test)
```

# Random Forests

```python
from sklearn.ensemble import RandomForestRegressor

regforest = RandomForestRegressor()
cvforest = GridSearchCV(regforest, param_grid = {'n_estimators' : range(100, 110)})
cvforest.fit(X_train, y_train.ravel())
cvforest.best_params_
```

```python
regforestotimo = RandomForestRegressor(n_estimators = 108)
regforestotimo.fit(X_train, y_train.ravel())
regforestotimo.score(X_test, y_test.ravel())
```

# OLS

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
reg.score(X_test, y_test)
```
